```python
from matplotlib import font_manager
def find_chinese_font():
    # Find a font that supports Chinese characters
    for font_path in font_manager.findSystemFonts(fontpaths=None, fontext='ttf'):
        try:
            font = font_manager.FontProperties(fname=font_path)
            sample_text = '测试中文字符'
            font_test = font_manager.FontProperties(fname=font_path).get_name()
            if font_test:
                return font_path
        except Exception as e:
            continue
    raise RuntimeError('>>>>>>>>>>>>>>>>>>>>>>>>>No suitable Chinese font found.')
```
